```python
features_to_drop = ['yearalter1', 'yearalter2']
categorical_features = [f for f in categorical_features if f not in features_to_drop ]
new_train_data = train_data[categorical_features + numerical_features]
```

```python
# mode imputation
categorical_data_mode = new_train_data[categorical_features].mode().iloc[0]
new_train_data.loc[:, categorical_features] = new_train_data[categorical_features].fillna(categorical_data_mode)
# mean imputation
numerical_data_mean = new_train_data[numerical_features].mean()
new_train_data.loc[:, numerical_features] = new_train_data[numerical_features].fillna(numerical_data_mean)
```

```python
for f in categorical_features:
    value_counts = new_train_data[f].value_counts().to_dict()
    if len(value_counts.values()) != len(set(value_counts.values())):
        print(f)
        print(value_counts)
        print()
```

```python
from category_encoders import MEstimateEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_train_data, train_data.target__office, test_size=0.33, stratify=train_data.target__office, random_state=42)
```

```python
one_hot_encode_columns = []
target_encode_columns = []
for f in categorical_features:
    unique_values = new_train_data[f].unique()
    if len(unique_values) > 5:
        target_encode_columns.append(f)
    else:
        one_hot_encode_columns.append(f)
        
one_hot_encode_columns
```

```python
ohe = OneHotEncoder(cols=one_hot_encode_columns, )
X_train_new = ohe.fit_transform(X_train, y_train)

mee = MEstimateEncoder(cols=target_encode_columns, m=2)
X_train_new = mee.fit_transform(X_train_new, y_train)
```

```python
X_test_new = ohe.transform(X_test, y_test)
X_test_new = mee.transform(X_test_new, y_test)
```

```python
from sklearn.feature_selection import mutual_info_classif
# determine the mutual information
mutual_info = mutual_info_classif(X_train_new, y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train_new.columns
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
from sklearn.feature_selection import SelectKBest
# Now we Will select the  top n important features
sel_cols = SelectKBest(mutual_info_classif, k=25)
sel_cols.fit(X_train_new, y_train)
X_train_new.columns[sel_cols.get_support()]
```

```python
selected_cols = X_train_new.columns[sel_cols.get_support()]
X_train_new = X_train_new[selected_cols]
X_test_new = X_test_new[selected_cols]
```

```python
num_true_records = sum(y_train == True) * 1
num_false_records = sum(y_train == False)
num_false_records
```

```python
num_classifiers = num_false_records // num_true_records
false_records_index = y_train.index[y_train == False].tolist()
true_records_index = y_train.index[y_train == True].tolist() 

classifiers = []
accuracies = []
for i in range(num_classifiers):
    from_ = i * num_true_records
    to_ = (i * num_true_records) + num_true_records
    false_train_index = false_records_index[from_: to_]
    
    X_ = X_train_new.loc[false_train_index + true_records_index]
    y_ = y_train.loc[false_train_index + true_records_index]
    idx = np.random.permutation(X_.index)
    X_.reindex(idx)
    y_.reindex(idx)
    clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
    clf.fit(X_, y_)
    classifiers.append(clf)
    
    y_hat = clf.predict(X_test_new)
    accuracies.append(accuracy_score(y_test.astype(int), y_hat.astype(int)))
    
    
    
```

```python
def predict(classifiers, X_test):
    f = lambda y: -1 if y == 0 else 1
    f_inv = lambda y: 0 if y < 0 else 1
    prediction = np.zeros(len(X_test))
    
    for clf in classifiers:
        y_hat = clf.predict(X_test).astype(int)
        prediction = prediction + np.array(list(map(f, y_hat)))
        
    return np.array(list(map(f_inv, prediction)))
    
y_hat = predict(classifiers, X_test_new)
```

```python
# 30 - Triple false
print(accuracy_score(y_test.astype(int), y_hat.astype(int)))
cm = confusion_matrix(y_test, y_hat)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```
